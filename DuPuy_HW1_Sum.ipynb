```python
# sum.ipynb
# Purpose: Compute sum of series that converges to pi
# Author: Jack DuPuy

import numpy as np
import matplotlib.pyplot as plt

# computes S(N) = sum_{j=1}^N 4/(2j-1) * (-1)^(j+1) using @, no for-loop or sum() needed
def sum_series(N):
    # no for loop needed because numpy can handle all iteration internally
    return ((-1)**(np.arange(1, N+1)+1)) @ (4/(2*np.arange(1, N+1)-1))

# plots S(n) vs n for n=1,...,N converging to pi using a for-loop
def plot(N):
    S_vals = []
    # use range (1, N+1) to get sum of numbers 1-50 instead of 0-49
    for n in range(1, N+1):
        S_vals.append(sum_series(n))

    n_vals = np.arange(1, N+1)

    plt.figure(figsize=(8,5))
    plt.plot(n_vals, S_vals, "o-", label="S(N)")
    plt.axhline(np.pi, color="red", linestyle="--", label="π")
    plt.xlabel("N")
    plt.ylabel("S(N)")
    plt.title("Convergence to π")
    plt.legend()
    plt.grid(True)
    plt.show()

# get user input for value N
# N = int(input("Enter N: "))
N = 50    # because problem statement asks for S(N) vs N for 1 <= N <= 50
plot(N)
```
