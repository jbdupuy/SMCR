```python
# fibo.ipynb
# Purpose: Generate Fibonacci sequences and illustrate convergence to the golden ratio
# Author: Jack DuPuy

import numpy as np
import matplotlib.pyplot as plt

# returns Fibonacci sequence of length N (N > 1) along with index array j = [1, 2, ..., N].
def seqnce(N):
    # (a) compute the Fibonacci sequence
    F = np.zeros(N, dtype=int)
    F[0], F[1] = 1, 1
    for k in range(2, N):
        F[k] = F[k-1] + F[k-2]

    # (b) compute the array of indices
    j = np.linspace(1, N, N, dtype=int)

    return j, F


# uses seqnce(N) to compute Fibonacci numbers and ratios then plots both the sequence and the convergence to the golden ratio.
def plot(N):
    # (a) get sequence and indices
    j, F = seqnce(N)

    # (b) compute ratio sequence F_j / F_{j+1}
    ratios = F[:-1] / F[1:]
    phi = (np.sqrt(5) - 1) / 2  # reciprocal golden ratio

    # (c) make plots
    fig, ax = plt.subplots(2, 1, figsize=(8, 6))

    # plot Fibonacci numbers
    ax[0].plot(j, F, "o-", label="Fibonacci numbers")
    ax[0].set_xlabel("Index j")
    ax[0].set_ylabel("F_j")
    ax[0].set_title("Fibonacci Sequence")
    ax[0].legend()
    ax[0].grid(True)

    # plot ratios converging to golden ratio
    ax[1].plot(j[:-1], ratios, "o-", label="F_j / F_{j+1}")
    ax[1].axhline(phi, color="red", linestyle="--", label="1/phi (limit)")
    ax[1].set_xlabel("Index j")
    ax[1].set_ylabel("Ratio F_j / F_{j+1}")
    ax[1].set_title("Convergence to Golden Ratio Reciprocal")
    ax[1].legend()
    ax[1].grid(True)

    plt.tight_layout()
    plt.show()

# get user input for value N
N = int(input("Enter N: "))
plot(N)
```
